```python
#working for Part 1-B:
#reading data from file
import glob

path = r'C:\Users\Atul\Downloads\Data2'  
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0,encoding='ISO-8859-1')
    df['State'] = filename.split('-')[1]
    li.append(df)

GSVA_df = pd.concat(li, axis=0, ignore_index=True,sort=False)
```

```python
#Checking the data
GSVA_df
```

```python
GSVA_df.info()
```

```python
#Dropping the unnecessary rows
GSVA_df1 = GSVA_df.drop(['2011-12','2012-13','2013-14','2015-16','2016-17'],axis  = 1)
```

```python
#getting information ablout my data
GSVA_df1.info()
```

```python
GSVA_df1
```

```python
#Calculating percentage for getting the Tertiary percentage
GSVA_df6['Tertiary_%'] = GSVA_df6['Tertiary']/GSVA_df6['Gross State Domestic Product']*100
```
